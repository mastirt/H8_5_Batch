```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split # data splitting
from sklearn.tree import DecisionTreeRegressor # ML model
from sklearn.metrics import mean_squared_error # evaluation

import category_encoders as ce
```

# V1

```python
df = sns.load_dataset('tips')
df
```

```python
df_encode = df[['sex', 'smoker', 'time']]
df_encode
```

```python
df_encode = pd.get_dummies(df_encode)
df_encode
```

```python
df_not_encode = df[['total_bill', 'tip', 'day', 'size']]
df_not_encode
```

```python
df = pd.concat([df_not_encode, df_encode], axis = 1)
df
```

```python
ordinal_encoder = ce.OrdinalEncoder(cols = ['day'])
df_day = ordinal_encoder.fit_transform(df['day'])
df_day
```

```python
df.drop(columns = 'day', inplace =True)
df
```

```python
df = pd.concat([df, df_day], axis = 1)
df
```

```python
X = df.drop(columns = 'tip')
y = df['tip']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size = 0.3)
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```

# V2

```python
df_v2 = sns.load_dataset('tips')
df_v2
```

```python
df_v2_encode = df_v2[['sex', 'smoker']]
df_v2_encode
```

```python
df_v2_encode = pd.get_dummies(df_v2_encode)
df_v2_encode
```

```python
df_v2_not_encode = df_v2[['total_bill', 'time','tip', 'day', 'size']]
df_v2_not_encode
```

```python
df_v2 = pd.concat([df_v2_not_encode, df_v2_encode], axis = 1)
df_v2
```

```python
binary_encoder = ce.OrdinalEncoder(cols = ['time','day'])
df_daytime = binary_encoder.fit_transform(df_v2[['time', 'day']])
df_daytime
```

```python
df_v2.drop(columns = 'day', inplace =True)
df_v2.drop(columns = 'time', inplace =True)
df_v2
```

```python
df_v2 = pd.concat([df_v2, df_daytime], axis = 1)
df_v2
```

```python
X_v2 = df_v2.drop(columns = 'tip')
y_v2 = df_v2['tip']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_v2, y_v2, random_state=10, test_size = 0.3)
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```
