```python
#Menggabungkan berdasarkan kolom year dan borough serta menjumlah valuenya
df_westminster = df_westminster.groupby(['year', 'borough'])['value'].sum().reset_index()
```

```python
#Menggabungkan berdasarkan kolom year dan borough serta menjumlah valuenya
df_greenwich = df_greenwich.groupby(['year', 'borough'])['value'].sum().reset_index()
```

```python
#Mengset Index dari tahun
df_greenwich.set_index('year', inplace=True)
df_greenwich
```

```python
#Menggabungkan berdasarkan kolom year dan borough serta menjumlah valuenya
df_lambeth = df_lambeth.groupby(['year', 'borough'])['value'].sum().reset_index()
```

```python
#Mengset Index dari tahun
df_lambeth.set_index('year', inplace=True)
df_lambeth
```

```python
#Menggabunkan data df_lambeth, df_greenwich dan df_westminster berdasarkan tahun 
df_gwl = pd.merge(df_lambeth, df_greenwich, on='year', how='inner')
df_gwl = pd.merge(df_gwl, df_westminster, on='year', how='inner')

df_gwl
```

```python
df_gwl
```

```python
#Membuat Diagram Line Plot
df_gwl.plot(kind='line')

#Membuat judul dan label pada Diagram
plt.title('Number of Crime Incidents in Westminster, Greenwich and Lambeth from 2008 to 2016')
plt.ylabel('Value of Incidents')
plt.xlabel('Years')

#Menampilkan Diagram
plt.show()
```

```python
#Membuat bin untuk histogram
count, bin_edges = np.histogram(df_hmb, 15)

#Membuat diagram histogram
df_hmb.plot(kind='hist',
          figsize=(10,6),
          bins=15,
          alpha=0.5,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'])

#Membuat judul dan histogram
plt.title('The Crime Count in Harrow, Merton, and Bexley from 2008 to 2016')
plt.xlabel('Value of Incidents')
plt.ylabel('year')

#Menampilkan Diagram
plt.show()
```
